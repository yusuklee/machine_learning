```python
from textwrap import indent

import torch
import torch.nn as nn




class LinearRegressionModel(nn.Module):
    def __init__(self,i,o):
        super().__init__()
        self.linear = nn.Linear(in_features=i,out_features=o)
        self.activation = nn.Sigmoid()

    def forward(self,x):
        return self.activation(self.linear(x))
```

```python
import torch.nn.functional as F
x= torch.ones(4)
y = torch.zeros(3)
model = LinearRegressionModel(4,3)

loss_function = nn.MSELoss()
```

```python
learning_rate = 0.01
nb_epochs = 1000
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)

for epoch in range(1001):
    y_pred = model(x)
    loss = loss_function(y_pred,y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
print(loss)
for param in model.parameters():
    print(param)
```

```python
import torch
import torch.nn as nn

class LinearRegressionModel(nn.Module):
    def __init__(self,i,o):
        super().__init__()
        self.linear1= nn.Linear(i,10)
        self.linear2 = nn.Linear(10,10)
        self.linear3 = nn.Linear(10,10)
        self.linear4 = nn.Linear(10,o)
        self.sibal = nn.LeakyReLU(0.1)

    def forward(self,x):
        hidden =self.sibal(self.linear1(x))
        hidden =self.sibal(self.linear2(hidden))
        hidden =self.sibal(self.linear3(hidden))
        y = self.linear4(hidden)
        return y
```

```python
x = torch.ones(4)
y = torch.zeros(3)
model = LinearRegressionModel(4,3)
loss_function = nn.MSELoss()
```

```python
learning_rate = 0.01
nb_epochs=1000
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)

for epoch in range(nb_epochs+1):
    y_pred = model(x)
    loss = loss_function(y_pred, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
print(loss)
for param in model.parameters():
    print(param)
```

```python
x = torch.ones(5000,10)
y = torch.zeros(5000,1)
learning_rate = 0.01
iter_num = 1000
minibatch_size = 256
```

```python
input_dim = x.size(-1)
output_dim = y.size(-1)

# 보통 hidden layer 는 출력에 가까울 수록 작아지게 설계하는 것이 일반적임 (더 좋은 성능)
model = nn.Sequential (
    nn.Linear(input_dim, 10),
    nn.LeakyReLU(0.1),
    nn.Linear(10, 8),
    nn.LeakyReLU(0.1),
    nn.Linear(8, 6),
    nn.LeakyReLU(0.1),
    nn.Linear(6, output_dim)
)

loss_function = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
indices = torch.randperm(x.size(0)) #(0 ~4999) 까지 섞은 배열을줌
print (indices)
x_batch_list = torch.index_select(x, 0, index=indices) # shuffle 된 데이터셋으로, 데이터양이 상당하므로, 미니배치 변수로 선언
y_batch_list = torch.index_select(y, 0, index=indices)
```

```python
for index in range(1000):
    indices = torch.randperm(x.size(0))

    x_batch_list = torch.index_select(x,0,index=indices)
    y_batch_list = torch.index_select(y,0,index=indices)
    x_batch_list = x_batch_list.split(minibatch_size,0)
    y_batch_list = y_batch_list.split(minibatch_size,0)

    for x, y in zip(x_batch_list, y_batch_list):
        y_pred = model(x)
        loss = loss_function(y_pred,y)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

print(loss)
for param in model.parameters():
    print(param)
```
