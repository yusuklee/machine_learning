```python
import torch
import torch.nn as nn

class NeuralNetwork(nn.Module):
    def __init__(self,input_size, output_size):

        super().__init__()

        self.W = torch.rand(input_size,output_size)
        self.b = torch.rand(output_size)


    def forward(self,x):
        return torch.matmul(x,self.W) +self.b
```

```python
x= torch.rand(4)
myLinear = NeuralNetwork(4,3)
y = myLinear(x)
print(y, y.shape)
```

```python
class sibal(nn.Module):
    def __init__(self, i,o):
        super().__init__()

        self.W = nn.Parameter(torch.rand(i, o))
        self.b = nn.Parameter(torch.rand(o))

    def forward(self,x):
        return torch.matmul(x,self.W)+self.b
```

```python
x= torch.FloatTensor([100,1000,2,3])
obj = sibal(4,100)
res = obj(x)
print(res,res.shape)
```
