```python
import torch
import torch.nn as nn
x = torch.ones(5000,10)
y= torch.zeros(5000,1)
lr = 0.01
nb =1000
mini = 256
```

```python
input_dim = x.size(-1)
output_dim = y.size(-1)

model = nn.Sequential(
    nn.Linear(input_dim,10),
    nn.ReLU(),
    nn.Linear(10,8),
    nn.ReLU(),
    nn.Linear(8,6),
    nn.ReLU(),
    nn.Linear(6,output_dim)
)

loss_function =nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters())
```

```python
indices = torch.randperm(x.size(0))
print(indices)
x_batch_list = torch.index_select(x,0,indices)
x_batch_list = x_batch_list.split(256,0)
y_batch_list = torch.index_select(y,0,indices)
y_batch_list = y_batch_list.split(256,0)
```

```python
for i in range(1000):
    indices = torch.randperm(x.size(0))

    x_batch_list = torch.index_select(x,0,indices)
    x_batch_list = x_batch_list.split(256,0)
    y_batch_list = torch.index_select(y,0,indices)
    y_batch_list = y_batch_list.split(256,0)

    for x,y in zip(x_batch_list,y_batch_list):
        y_pred = model(x)
        loss = loss_function(y_pred,y)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

print(loss)
for param in model.parameters():
    print(param)
```
